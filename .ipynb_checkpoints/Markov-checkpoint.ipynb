```python
import numpy as np
import numpy.random as rd
from collections import Counter
import matplotlib.pyplot as plt
```

# Slow but easy to understand method 1

```python
def Chain(x_0, p_hop, n_its):
    p_vec = np.zeros([n_its, 6])
    N_paths = len(x_0)
    for i in np.arange(n_its):
        for j in np.arange(len(x_0)):
            p = rd.rand()
            if p <= p_hop:
                x_0[j] = (x_0[j] - 1) % 6
            elif p >= (1-p_hop):
                x_0[j] = (x_0[j] + 1) % 6
        for key, value in Counter(x_0).items():
            p_vec[i,key] = value / N_paths
    return x_0, p_vec
```

```python
x, p = Chain(x_0=[1]*10000, p_hop=0.001, n_its=50)
```

```python
for pos in np.arange(5):
    plt.plot(np.arange(len(p[:])), p[:,pos], label = pos if pos != 0 else 6)
    plt.legend()
```

# Fast, but more complex method

```python
def Chain_fast(x_0, p_hop, n_its):
    rng = np.random.default_rng(seed=5)
    p_vec = np.zeros([n_its, 6])
    N_paths = len(x_0)
    
    for i in np.arange(n_its):
        random = rng.random(N_paths)
        x_0 = (np.where(random < p_hop, x_0+1, np.where(random > (1-p_hop), x_0-1, x_0))) % 6
        p_vec[i] =  np.bincount(x_0.astype(int), minlength=6) / N_paths
    return x_0, p_vec
```

```python
x, p = Chain_fast(x_0=np.ones(1000), p_hop=0.01, n_its=200)
```

```python
for pos in np.arange(5):
    plt.plot(np.arange(len(p[:])), p[:,pos], label = pos if pos != 0 else 6)
    plt.legend()
```
